```python
%%manim -qm -v WARNING TableExamples

class TableExamples(Scene):
    def construct(self):
        t0 = Table(
            [["True", "False"],
            ["False", "True"]],
            col_labels=[MathTex(r"P", font_size=96), MathTex(r"\neg P", font_size=96)]).set_column_colors(WHITE, YELLOW)
        self.add(t0)
        self.wait()
        self.remove(t0)
        
        t1 = Table(
            [["True", "True", "True"],
            ["False", "True", "False"],
            ["True", "False", "False"],
            ["False", "False", "False"]],
            col_labels=[MathTex(r"P", font_size=96), MathTex(r"Q", font_size=96),
                        MathTex(r"P\wedge Q", font_size=96)]).set_column_colors(WHITE, WHITE, YELLOW)
        self.add(t1)
        self.wait()
        self.remove(t1)
        
        t2 = Table(
            [["True", "True", "True"],
            ["False", "True", "True"],
            ["True", "False", "True"],
            ["False", "False", "False"]],
            col_labels=[MathTex(r"P", font_size=96), MathTex(r"Q", font_size=96),
                        MathTex(r"P\vee Q", font_size=96)]).set_column_colors(WHITE, WHITE, YELLOW)
        self.add(t2)
        self.wait()
        self.remove(t2)
        
        t3 = Table(
            [["True", "True", "True"],
            ["False", "True", "True"],
            ["True", "False", "False"],
            ["False", "False", "True"]],
            col_labels=[MathTex(r"P", font_size=96), MathTex(r"Q", font_size=96),
                        MathTex(r"P\Rightarrow Q", font_size=96)]).set_column_colors(WHITE, WHITE, YELLOW)
        self.add(t3)
        self.wait()
        self.remove(t3)
        
        t4 = Table(
            [["True", "True", "True"],
            ["False", "True", "False"],
            ["True", "False", "False"],
            ["False", "False", "True"]],
            col_labels=[MathTex(r"P", font_size=96), MathTex(r"Q", font_size=96),
                        MathTex(r"P\Leftrightarrow Q", font_size=96)]).set_column_colors(WHITE, WHITE, YELLOW)
        self.add(t4)
        self.wait()
        self.remove(t4)
        
        title = Text("Applications", weight=BOLD, font_size=108)
        self.play(Write(title))
        self.wait()
```

```python
%%manim -qm -v WARNING AddPackageLatex

class AddPackageLatex(Scene):
    def construct(self):
        myTemplate = TexTemplate()
        myTemplate.add_to_preamble(r"\usepackage{mathrsfs}")
        tex = Tex(
            r"$\mathscr{H} \rightarrow \mathbb{H}$}",
            tex_template=myTemplate,
            font_size=144,
        )
        self.add(tex)
```
